```python
# Data handling
import numpy as np
import pandas as pd

# Data visulazation
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Model selection and optimization
from sklearn import datasets, neighbors
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from collections import Counter
```

```python
# Dataset read in array form
data = np.genfromtxt("/home/shri/Music/Applied-AI-Course-master/KNN(K-NearestNeighbour)_Classification_n_Regression/demo_data/6.overlap.csv", delimiter=',')
```

```python

def knn_comparision(data, n_neighbors):
    """
    Data read, model_build, scatter plot
    """
    x = data[:, :2]

    #data[:,2] indicates col 3
    y = data[:, 2]

    #grid cell size
    h= 0.02

    # mesh color light red and light blue
    cmap_light = ListedColormap(['#FFAAAA', '#AAAAFF'])

    # datapoints colors Dark Red and Dark Blue
    cmap_bold = ListedColormap(['#FF0000','#0000FF'])

    clf = neighbors.KNeighborsClassifier(n_neighbors)#n_neighbors
    clf.fit(x,y)

    # indicates min and max of 1st col
    x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1

    #indicates min and max of 2nd col
    y_min ,y_max = x[:, 1].min() - 1, x[:, 1].max() + 1

    # we create a mesh grid (x_min,y_min) to (x_max,y_max) with 0.02 grid spaces
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))

    # we predict the value (either 0 or 1) for each element in the grid
    z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    #convert back the xx to its original shape (we need it to plot the decission boundry)
    z = z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, z, cmap=cmap_light)

    # scatter plot of with given points
    plt.scatter(x[:, 0], x[:, 1], c=y, cmap=cmap_bold)

    #defining scale on both axises
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())

    # set the title
    plt.title('K value = '+str(n_neighbors))#n_neighbors
    plt.show()

```

```python
knn_comparision(data, 1)
```

```python
knn_comparision(data, 5)
```
